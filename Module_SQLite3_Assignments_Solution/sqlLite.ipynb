```python
#  connect to an sqlite db

import sqlite3
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()

# create a table using cursor object

cursor.execute('''
Create Table If Not Exists employees(
            id Integer Primary Key,
            name Text Not Null,
            age Integer,
            department text
        )

''')

#commit the changes
connection.commit()
```

```python
cursor.execute('''
Select * from employees

''')
```
