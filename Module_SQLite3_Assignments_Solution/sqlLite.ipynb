```python
#  connect to an sqlite db

import sqlite3
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()

# create a table using cursor object

cursor.execute('''
Create Table If Not Exists employees(
            id Integer Primary Key,
            name Text Not Null,
            age Integer,
            department text
        )

''')

#commit the changes
connection.commit()
```

```python
cursor.execute('''
Select * from employees

''')
```

```python
from os import name
# Insert the data in sqlite table

cursor.execute('''
Insert Into employees(name, age, department)
            values('shubair', 21, 'machine learning engineer')
''')

cursor.execute('''
Insert Into employees(name, age, department)
            values('sahil', 14, 'deep learning engineer')
''')

cursor.execute('''
Insert Into employees(name, age, department)
            values('zaman', 18, 'robotics engineer')
''')

connection.commit()
```

```python
# now execute and quering the data from the table to fetch the data from database

cursor.execute('Select * from employees')

rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
# update the data in the table of database

cursor.execute('''
UPDATE employees
Set age=34
where name='shubair'

''')

connection.commit()
```

```python
# now fetching again the data to see the changes

cursor.execute('Select * from employees')

rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
# delete the data

cursor.execute('''

Delete from employees
where name='sahil'

''')

cursor.execute('Select * from employees')

rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
# Pehle ek sales table banayenge
cursor.execute('''
Create Table If Not Exists sales(
    id Integer Primary Key,
    date Text,
    product_name Text,
    quantity Integer,
    region Text
)
''')

# Sales data array (List of Tuples)
# Format: (Date, Product Name, Sales Quantity, Region)
sales_data = [
    ('2024-05-01', 'Laptop', 15, 'North'),
    ('2024-05-02', 'Smartphone', 40, 'South'),
    ('2024-05-03', 'Wireless Mouse', 100, 'East'),
    ('2024-05-04', 'Mechanical Keyboard', 25, 'West'),
    ('2024-05-05', 'Monitor 24-inch', 10, 'North'),
    ('2024-05-06', 'Headphones', 50, 'South')
]


# Ek hi line mein saara data insert karne ke liye executemany() use karein
cursor.executemany('''
Insert Into sales(date, product_name, quantity, region)
            values(?, ?, ?, ?)
''', sales_data)

connection.commit()
```

```python
cursor.execute('SELECT * FROM sales')

rows = cursor.fetchall()

for row in rows:
    print(row)
```
