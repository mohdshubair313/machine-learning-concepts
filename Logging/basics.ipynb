```python
# we are gonna learn about logging in python

import logging

# configure the basic logging settings

logging.basicConfig(level=logging.DEBUG)

# log messages with different severity levels

logging.debug("This is a debug message")  # Detailed info, typically for diagnosing problems (lowest severity)
logging.info("This is an info message")  # Confirmation that things are working as expected

logging.warning("This is a warning message")  # Indication of something unexpected or potential problem
logging.error("This is an error message")  # Due to a more serious problem, the software couldn't perform some function
logging.critical("This is a critical message")  # A serious error, indicating the program may be unable to continue running (highest severity)
```

## Multiple Loggers

**Why use multiple loggers?**
- Different parts of your app may need different log levels or outputs
- You can control each module's logging independently
- Example: DB logs go to a file, UI logs go to console

**Use Cases:**
- `logger_db` → logs database queries to `db.log`, level=INFO
- `logger_network` → logs network calls to `network.log`, level=WARNING
- `logger_ui` → logs user actions to console, level=DEBUG

```python
import logging

# Create named loggers (isolated from each other)
logger_db = logging.getLogger("database")
logger_network = logging.getLogger("network")
logger_ui = logging.getLogger("ui")

# Set different levels for each
logger_db.setLevel(logging.INFO)
logger_network.setLevel(logging.WARNING)
logger_ui.setLevel(logging.DEBUG)

# Create handlers (where logs go)
console_handler = logging.StreamHandler()  # prints to console
file_handler = logging.FileHandler("app.log")  # writes to file

# Create formatter
formatter = logging.Formatter("%(name)s - %(levelname)s - %(message)s")
console_handler.setFormatter(formatter)
file_handler.setFormatter(formatter)

# Attach handlers to specific loggers
logger_db.addHandler(file_handler)  # DB logs go to file
logger_network.addHandler(file_handler)  # Network logs also to file
logger_ui.addHandler(console_handler)  # UI logs to console

# Example usage
logger_db.info("Connected to PostgreSQL successfully")  # will be logged to file
logger_network.warning("API request timed out after 5s")  # will be logged to file
logger_ui.debug("User clicked button 'Submit'")  # will print to console
logger_ui.info("User logged in")  # will print to console
```
