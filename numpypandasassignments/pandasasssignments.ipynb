### Assignment 4: Data Aggregation

1. Create a Pandas DataFrame with 2 columns: 'Category' and 'Value'. Fill the 'Category' column with random categories ('A', 'B', 'C') and the 'Value' column with random integers. Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category.
2. Create a Pandas DataFrame with 3 columns: 'Product', 'Category', and 'Sales'. Fill the DataFrame with random data. Group the DataFrame by 'Category' and compute the total sales for each category

```python
# Q1 - Hint for assignment 4 ==>  Jab bhi "har category ke liye", "har department ke liye", ya "for each" type ka sawaal aaye, wahan aankh band karke groupby() lagana hota hai!

category = np.random.choice(['A','B','C'], size=10)

values = np.random.randint(10,100,size=10)

df = pd.DataFrame({"category":category, 'value': values})
print(df)

# Humne 'Category' se group kiya, 'Value' column select kiya, aur agg() kiya
# groupby('Category') ne saare 'A' ko alag, 'B' ko alag aur 'C' ko alag group kiya.
# ['Value'] ne bataya ki calculation 'Value' column par karni hai.
# .agg(['mean', 'sum']) ne har category ka average (mean) aur total (sum) dono ek hi baar mein nikal diya!

result = df.groupby('category')['value'].agg(['mean','sum'])
print(result)
```

1. Create a Pandas DataFrame with a MultiIndex (hierarchical index). Perform some basic indexing and slicing operations on the MultiIndex DataFrame.
2. Create a Pandas DataFrame with MultiIndex consisting of 'Category' and 'SubCategory'. Fill the DataFrame with random data and compute the sum of values for each 'Category' and 'SubCategory'.

```python
# Normal DataFrame mein har row ka sirf ek address (Index) hota hai. Par MultiIndex mein har row ka address multiple parts se mil kar banta hai.

# School ka sabse best example: Agar school mein sirf Roll No ko index banayein:

# Roll No 1
# Roll No 2
# Roll No 1 (Ye doosri class ka ho sakta hai, confusion!)
# Toh hum MultiIndex banate hain:

# Level 1: Class
# Level 2: Roll No
# Address ab aisa dikhega:

# ('10th', 1)
# ('10th', 2)
# ('11th', 1)
# Yahan ek se zyada index (Class aur Roll No) milkar row ko identify kar rahe hain. ISI KO MULTI-INDEX BOLTE HAIN!


# we are making multi index with the array or lists
outerarr = ['A','B','C']
innerarr = ['shubair','sahil','shubham']


multi = pd.MultiIndex.from_arrays([outerarr, innerarr], names=['category','subcategory'])

dataframe = pd.DataFrame(np.random.randint(1,10, size=(3,3)), columns=['A', 'B','C'], index=multi)

# Ab dataframe se category 'A' select karo
multi_df = dataframe.loc['A']
print(multi_df)

```

```python
categories = ['furniture', 'electronics', 'clothing']
subcategory = ['table', 'tv', 'shirt']

multi_index = pd.MultiIndex.from_arrays([categories, subcategory], names=['category', 'subcategory'])

df = pd.DataFrame(np.random.randint(1,10, size=(3,1)), columns=['quantity'], index=multi_index)

print(df)

result = df.groupby(['category', 'subcategory']).sum()

print(result)
```

### Assignment 8: Pivot Tables

1. Create a Pandas DataFrame with columns 'Date', 'Category', and 'Value'. Create a pivot table to compute the sum of 'Value' for each 'Category' by 'Date'.
2. Create a Pandas DataFrame with columns 'Year', 'Quarter', and 'Revenue'. Create a pivot table to compute the mean 'Revenue' for each 'Quarter' by 'Year'.


```python
# Q1

dat_range = pd.date_range(start='2022-01-11', end='2022-01-15', freq='D') # Choti date range taaki output clean dikhe
df = pd.DataFrame({'Date': dat_range})

df['category'] = np.random.choice(['A','B','C'], size=len(dat_range))
df['value'] = np.random.randint(1,100, size=len(dat_range))

print("--- Original DataFrame ---")
print(df)
print("\n")
# 3. Pivot Table banayi
# Humne columns ke naam string mein pass kiye hain!
pivot_df = df.pivot_table(values='value', index='Date', columns='category', aggfunc='sum')
print("--- Pivoted Table ---")
print(pivot_df)
```

```python
# Q2

# 1. Size choose karo (jaise 12 ya 15 entries)
n = 15
# 2. Random choices se values generate karo
years_list = np.random.choice([2021, 2022, 2023], size=n)
quarters_list = np.random.choice(['Q1', 'Q2', 'Q3', 'Q4'], size=n)
revenue_list = np.random.randint(100, 1000, size=n)
# 3. Ek single DataFrame banao
df = pd.DataFrame({
    'Year': years_list,
    'Quarter': quarters_list,
    'Revenue': revenue_list
})

# 4. Pivot table banao
pt = df.pivot_table(
    values='Revenue', 
    index='Year', 
    columns='Quarter', 
    aggfunc='mean'   # Question ne mean bola hai!
)

print("--- Pivot Table of Average Revenue ---")
print(pt)
```

### Assignment 9: Applying Functions

1. Create a Pandas DataFrame with 3 columns and 5 rows filled with random integers. Apply a function that doubles the values of the DataFrame.
2. Create a Pandas DataFrame with 3 columns and 6 rows filled with random integers. Apply a lambda function to create a new column that is the sum of the existing columns.

```python
#Q1

df = pd.DataFrame(np.random.randint(1, 16, size=(5,3)), columns=['A', 'B','C'])

print(df)

# now we apply some function on this dataframe
# for example we want to apply lambda function on this dataframe which will add 10 to each

df = df.apply(lambda x: x * 2)

print("this is the new version of df where we apply the function on each of vlaues in dataframe: ", df)
```

```python
df = pd.DataFrame(np.random.randint(1, 19, size=(6,3)), columns=['A', 'B','C'])

print(df)

# create a lambda function which is sum of existing columns and return that column as new column

df['D'] = df.apply(lambda x: x['A'] + x['B'] + x['C'], axis=1)
print(df)

```
