```python
# 1. Create a NumPy array of shape (5, 5) filled with random integers between 1 and 20. Replace all the elements in the third column with 1.

import numpy as np

# step 1- 5x5 ka ek random arr banate hai jisme 1 se 20 tak numbers honge
arr = np.random.randint(1,21,size=(5,5))

print("--- original array ---- ")
print(arr)

# step 2: Teesre column (Index 2) ke saare elements ko 1 se replace karte hain
# [:, 2] ka matlab -> ":" yani saari rows, aur "2" yani teesra column

# Numpy mein kisi hisse ko select karne ka tarika hota hai: [rows, columns]. Humein saari rows chahiye, isliye hum sirf colon : lagayenge. Aur humein teesra column chahiye, isliye hum index 2 lagayenge (Kyunki ginti 0 se shuru hoti hai: 0, 1, 2). arr[:, 2] = 1

arr[:, 2] = 1

print("\n--- Naya Array (3rd column ko 1 karne ke baad) ---")
print(arr)
```

```python
# ### Assignment 2: Array Indexing and Slicing 
# 1. Create a NumPy array of shape (6, 6) with values from 1 to 36. Extract the sub-array consisting of the 3rd to 5th rows and 2nd to 4th columns.

numArr = np.arange(1,37).reshape((6,6))
slicingArr = numArr[2:5, 1:4]
print("-------------")
print(slicingArr)
print(numArr)
```

```python
# 2. Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border.

arr_1 = np.random.randint(1,26, size=(5,5))
print(arr_1)

top_row = arr_1[0,:]
bottom_row = arr_1[-1,:]
left_border= arr_1[1:-1, 0] # (Matlab row 1 se leke aakhiri ke pehle tak, aur 0th column)
right_border = arr_1[1:-1, -1]

borders = np.concatenate((top_row, bottom_row, left_border, right_border))

print('\nBorders:')
print(borders)
```

```python
arr = np.random.randint(1,17,size=(4,4))
print(arr)
print("\n")

print("----- ye row wise addition ------")
row_wise_sum = np.sum(arr, axis=1)
print(row_wise_sum)

print("\n")

print("---------ye col wise addition hai ---------")
col_wise_sum = np.sum(arr, axis=0)

print(col_wise_sum)

```
