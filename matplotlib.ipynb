```python
# Cheatsheet

# ==========================================
# 📊 MATPLOTLIB - INDUSTRY MOST USED FUNCTIONS
# ==========================================
# Matplotlib mainly uses the 'pyplot' module (imported as plt).
import matplotlib.pyplot as plt

# --- 1. Plotting Types (Graphs) ---
# plt.plot(x, y)       -> Line chart (Time/Years ke trend dekhne ke liye)
# plt.scatter(x, y)    -> Scatter plot (Do cheezon mein relation nikalne ke liye)
# plt.bar(x, y)        -> Bar chart (Categories compare karne ke liye)
# plt.hist(x)          -> Histogram (Dekhne ke liye ki data kaise spread hai)
# plt.pie(x)           -> Pie chart (Percentages dikhane ke liye)

# --- 2. Customization (Sajawat) ---
# plt.title("My Graph")     -> Graph ka title
# plt.xlabel("X-Axis Name") -> Niche wale axis ka naam
# plt.ylabel("Y-Axis Name") -> Side wale axis ka naam
# plt.legend()              -> Side mein box banata hai ki kaunsa color kiska hai
# plt.figure(figsize=(8,4)) -> Graph paper ka size set karta hai (Width, Height)
# plt.grid(True)            -> Background mein grid (lines) dikhata hai

# --- 3. Structural ---
# plt.subplot(1, 2, 1)      -> Multiple graphs ke liye grid banata hai (Rows, Cols, Position)
# plt.show()                -> Akhir mein graph screen par print karta hai
```

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#create a line plot

plt.plot(x,y)

# you can give extra details on graph like x/y axis name, title and more
plt.xlabel('Eggs in india')
plt.ylabel('People in india')
plt.title("basic graph for showing how eggs available in india")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#customize the line plot

plt.plot(x,y,color='black',linestyle='dashdot', marker='o', linewidth=3)
plt.grid(True)
```

```python
x = [1,2,3,4,5]
y1=[4,8,12,16,20]
y2=[12,24,36,48,60]

plt.figure(figsize=(9,5)) # figure size or chart size you are mentioning like in pixels

# --- SUBPLOT KA FUNDA ---
# Syntax: plt.subplot(Rows, Columns, Position Number)
# Hamein shuru mein hi bata dena hota hai ki hamari grid ka naksha kaisa hoga.
# Niche ke code mein hum Python ko bata rahe hain: "Bhai hum 1 Row aur 3 Columns (Total 3 dibbe) banane wale hain."
# --- PEHLA DIBBA (1st Graph) ---
# (Naksha: 1 Row, 3 Columns ki grid. Abhi drawing 1st number ke dibbe mein chal rahi hai)
plt.subplot(1, 3, 1) 
plt.plot([1, 2, 3], [1, 4, 9], color='red')
plt.title("Graph 1")
# --- DUSRA DIBBA (2nd Graph) ---
# (Naksha: Wahi same 1x3 ka. Par ab drawing 2nd number ke dibbe mein karni hai)
plt.subplot(1, 3, 2) 
plt.plot([1, 2, 3], [1, 8, 27], color='blue')
plt.title("Graph 2")
# --- TEESRA DIBBA (3rd Graph) ---
# (Naksha: Wahi same 1x3 ka. Par ab drawing 3rd number ke dibbe mein karni hai)
# NOTE: Agar aap yahan "plt.subplot(1, 2, 3)" likhte toh error aata kyunki 1x2 ki grid mein sirf 2 hi dibbe ban sakte hain, teesra nahi!
plt.subplot(1, 3, 3) 
plt.plot([1, 2, 3], [1, 16, 81], color='green')
plt.title("Graph 3")
# Jab saari drawings complete ho jayein, toh display karo
plt.show()
```

```python
# Bar graph or plot

categories = ['A','B','C','F','E']

values=[1,2,3,4,5]

# create a bar plot

plt.bar(categories,values, color='purple')
plt.xlabel('Categories')
plt.ylabel('values')
plt.title('Bar plot')
plt.show()
```

```python
# Histograms are sued to represent the distribution of a dataset.
# They divide data into bins and count numbers of data points in each bin
```

```python
data = [1,2,3,3,4,5,6,7,7]

plt.hist(data,bins=5, color='orange', edgecolor="black")
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.scatter(x,y,color="blue", marker='x')
```

```python
# pie chart 

labels = ['A', 'B', 'C', 'D']
Sizes = [30,20,12,24]

colors=['gold', 'silver', 'red', 'green']

plt.pie(Sizes, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True)

# Note: if you want to mention percentage then use --> autopct which shows automatically percentage of a chart
# if you want to explode or detach a section of chart and show in picture then use shadow
```
